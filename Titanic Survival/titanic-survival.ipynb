```python
df = pd.read_csv("/kaggle/input/titanic/train.csv")
df.head()
```

```python
df = df.drop(["Name","Ticket","Age","Embarked","Fare","Cabin"],axis=1)
df
```

```python
df1 = pd.get_dummies(df, columns=['Sex'], dtype=int)
df1=df1.drop(["Sex_female"],axis=1)
df1
```

```python
df1.dropna(inplace=True)
```

```python

X = df1.drop(["Survived"],axis =1)
y = df1["Survived"]
X
```

```python
from sklearn.ensemble import RandomForestClassifier
model = RandomForestClassifier(n_estimators=10, max_depth=5, random_state=42)
model.fit(X, y)
```

```python
dft = pd.read_csv("/kaggle/input/titanic/test.csv")
dft
```

```python

dft = dft.drop(["Name","Ticket","Age","Fare","Embarked","Cabin"],axis=1)
df2 = pd.get_dummies(dft, columns=['Sex'], dtype=int)
df2=df2.drop(["Sex_female"],axis=1)
df2
```

```python
df2.isnull().sum()
```

```python
predictions = model.predict(df2)
predictions
```

```python
output = pd.DataFrame({'PassengerId': df2.PassengerId, 'Survived': predictions})
output.to_csv('submission.csv', index=False)
```
